#### 3.1

Let $X(t)$ with  $t > 0$  be a brownian motion process with a drift parameter $\mu$ and variance $\sigma^2$ for which $X(0) = 0$ 

Show that $-X(t), t \ge 0$ is a brownian motion process with drift parameter $- \mu$ and variance parameter $\sigma^2$

Solution: 

Given that $X(t)$ is a brownian motion process, we know $X(t+y) - X(y) \sim N(\mu t, \sigma^2 t)$ 

Thus $-X(t) = -(X(t+y) - X(y)) = -X(t+y) + X(y)$ 

Because $X(0) = 0$ 

$-X(t+0) + 0 = -X(t)$ 

$\Rightarrow E[-X(t)] = -E[X(t)] = - \mu$


`as seen on page 36`

$Var(-(X(t) + X(0)) = Var(-X(t)) =  var(\sigma\sqrt{\Delta} \Sigma_{i = 1}^{\frac{t}{\Delta}} - X_i) $

$= var( - \sigma\sqrt{\Delta} \Sigma_{i = 1}^{\frac{t}{\Delta}}  X_i)$

$ = -\sigma^2 \Delta \Sigma_{i =1}^{\frac{t}{\Delta}} var(X_i)$ `by independence`

$ = -\sigma^2 t [1 - (2p -1)^2]$

Because $p \rightarrow \frac{1}{2} as $\Delta \rightarrow 0$, the preceding shows that 

$Var(-X(t)) \rightarrow -t\sigma^2$

#### 3.4a 

Let $S(t) , t \ge 0$  be a geometric brownian motion process with $\mu = 0.1$ and volatility parameter $\sigma = 0.2$ 

Find $P(S(1) > S(0)$ 

#### Solution: 

Assume $S(0) \neq 0$ 

Then $P(S(1) > S(0)) = P(\frac{S(1+0)}{S(0)} > 1)$ 

$= P(log(\frac{S(1+0)}{S(0)}) > 0)$  `log(1) = 0 when base is e`

But $log(\frac{S(1+0)}{S(0)}) \sim N(0.1(1), (1)(0.2)^2)$ 

$ = log(\frac{S(1+0)}{S(0)}) \sim N(0.1, 0.04)$


So now $P(log(\frac{S(1+0)}{S(0)}) > 0)$ 

$ = P(Z > \frac{0 - 0.1}{0.2})$ 

$ = P(Z > - 0.5)$

$ = P(Z < 0.5)$ 

$ = 0.6915$ 
 
#### 3.4c 

Find $P(S(2) > S(1) > S(0)$ 

#### Solution 

$P(S(2) > S(1) > S(0) \Rightarrow P(S(2) > S(1) and P(S(1) > S(2))$ 

Note $S(2) - S(1)$ and $(S(1) - S(0))$ are i.i.d because they don't have overlapping time periods.  Then that also means they are RVs that come from identical distributions. 

Thus $P(S(2) > S(1)) = P(\frac{S(2+1)}{s(1)} > 1)$

$ = P(log(\frac{S(2+1)}{s(1)}) > 0)$

fact: $log(\frac{S(2+1)}{s(1)}) \sim N( \mu(1), (1)\sigma^2)$

thus 

$ = P(Z > \frac{0 - 0.1}{0.2})$ 

$ = P(Z > - 0.5)$

$ = P(Z < 0.5)$ 

$ = 0.6915$ 

So multiply two iid probabilities $0.6915 \cdot 0.6915$ to get 

$ = 0.4782$
